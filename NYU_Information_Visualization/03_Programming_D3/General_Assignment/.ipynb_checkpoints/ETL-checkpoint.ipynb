```python
import pandas as pd
```

```python
routes = pd.read_csv(r'routes.csv')
routes.head(3)
```

```python
def countRoutes(group):
    ''' Counts the number of flight per Airline and stores it in a new column
        Input: groupby broup
    '''
    count = group['AirlineID'].count()
    group['Count'] = count
    
    return group
```

```python
# Run the countRoutes function on the group
df = routes.groupby("AirlineID").apply(countRoutes)

# Drops the duplicates on AirlineID
df = df.drop_duplicates(['AirlineID'])

# Subset and sort the df
df = df[['AirlineID', 'AirlineName', 'Count']]
df = df.sort_values(['Count'], ascending=False).reset_index(drop=True)
df
```
